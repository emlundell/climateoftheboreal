```python
import graphql as gq
import pandas as pd

import altair as alt
import altair_recipes as ar
alt.renderers.enable('notebook')
alt.data_transformers.enable('json')
```

```python
query = """
{
  levels(order_by: {header: {year: asc, month: asc, day: asc, hour: asc}}, where: { lvltyp1: {_eq: 1}}) {
    wdir
    wspd
    header {
      year
      month
      day
      hour
    }
  }
}
"""
```

```python
df = gq.query(query, top='levels')
```

```python
df['wspd'] = df['wspd']/10.
df = df.rename({'header.year': 'year', 'header.month': 'month', 'header.day': 'day', 'header.hour': 'hour'}, axis=1)
yearly = df.groupby('year', as_index=False)
yearly = yearly.head(10000)
```

```python
base = alt.Chart(yearly).encode(
    x='year:O', 
    y=alt.Y('mean(wspd):Q')
)

errors = alt.Chart(yearly).mark_errorbar(extent='stdev').encode(
    x='year:O',
    y=alt.Y('wspd:Q', scale=alt.Scale(zero=False))
)

min_ = alt.Chart(yearly).mark_line(color='green').encode(
    x='year:O',
    y=alt.Y('min(wspd):Q')
)

max_ = alt.Chart(yearly).mark_line(color='green').encode(
    x='year:O',
    y=alt.Y('max(wspd):Q')
)
```

```python
graph = errors + base.mark_line() + base.mark_point() + min_ + max_
```
